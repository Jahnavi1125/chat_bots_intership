```python
history = []

def text_chat(prompt):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    answer = response.text

    history.append({
        "role": "User",
        "message": prompt
    })

    history.append({
        "role": "Assistant",
        "message": answer
    })

    return answer
```

```python
print(text_chat("What is Machine Learning?"))
```

```python
def image_chat(image, question):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[
            image,
            question
        ]
    )

    return response.text
```

```python
print(image_chat(image, "What objects are present in this image?"))
```

```python
history = []

def text_chat(prompt):

    history.append({
        "role": "User",
        "message": prompt
    })

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    answer = response.text

    history.append({
        "role": "Assistant",
        "message": answer
    })

    return answer
```

```python
print(text_chat("What is AI?"))

print(text_chat("Explain Deep Learning."))

print(text_chat("Difference between AI and ML"))
```

```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

print("Model Loaded Successfully!")
```
