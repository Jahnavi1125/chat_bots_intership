```python
import requests
import chromadb
from bs4 import BeautifulSoup
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

print("Model Loaded")
```

```python
client = chromadb.Client()

collection = client.get_or_create_collection(
    name="knowledge"
)

print("Database Created")
```

```python
def fetch_text(url):

    page = requests.get(url)

    soup = BeautifulSoup(
        page.text,
        "html.parser"
    )

    return soup.get_text()
```

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"

text = fetch_text(url)

chunks = [
    text[i:i+500]
    for i in range(0, len(text), 500)
]

for i, chunk in enumerate(chunks):

    embedding = model.encode(
        chunk
    ).tolist()

    collection.add(
        ids=[str(i)],
        documents=[chunk],
        embeddings=[embedding]
    )

print("Knowledge Added")
```

```python
def search(query):

    query_embedding = model.encode(
        query
    ).tolist()

    results = collection.query(
        query_embeddings=[
            query_embedding
        ],
        n_results=3
    )

    return results
```

```python
results = search(
    "What is Artificial Intelligence?"
)

for doc in results["documents"][0]:

    print(doc)
    print("\n")
```

```python
def update_knowledge(url):

    text = fetch_text(url)

    chunks = [
        text[i:i+500]
        for i in range(
            0,
            len(text),
            500
        )
    ]

    start_id = collection.count()

    for i, chunk in enumerate(chunks):

        embedding = model.encode(
            chunk
        ).tolist()

        collection.add(
            ids=[
                str(start_id+i)
            ],
            documents=[chunk],
            embeddings=[embedding]
        )

    print(
        "Database Updated"
    )
```

```python
update_knowledge(
    "https://en.wikipedia.org/wiki/Machine_learning"
)
```

```python
results = search(
    "What is Machine Learning?"
)

for doc in results["documents"][0]:

    print(doc)
    print("\n")
```

```python
import requests
from bs4 import BeautifulSoup

def fetch_text(url):

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(
        url,
        headers=headers,
        timeout=20
    )

    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    paragraphs = soup.find_all("p")

    text = ""

    for p in paragraphs:
        text += p.get_text() + "\n"

    return text
```

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"

text = fetch_text(url)

print(text[:2000])
```

```python
def ask_chatbot(question):

    query_embedding = model.encode(
        question
    ).tolist()

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=3
    )

    answer = "\n\n".join(
        results["documents"][0]
    )

    return answer
```

```python
print(
    ask_chatbot(
        "What is Artificial Intelligence?"
    )
)
```

```python
from bs4 import BeautifulSoup

with open("/content/AI_page.html", "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")

# Remove unwanted tags
for tag in soup(["script", "style", "nav", "header", "footer", "aside"]):
    tag.decompose()

text = "\n".join(
    p.get_text(strip=True)
    for p in soup.find_all("p")
    if len(p.get_text(strip=True)) > 30
)

with open("Artificial_Intelligence.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("Text file created successfully!")
```
