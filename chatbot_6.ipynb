```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

model_name = "facebook/nllb-200-distilled-600M"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)

print("Translation model loaded successfully!")
```

```python
def translate(text, src_lang, tgt_lang):

    tokenizer.src_lang = src_lang

    inputs = tokenizer(
        text,
        return_tensors="pt"
    ).to(device)

    translated_tokens = model.generate(
        **inputs,
        forced_bos_token_id=tokenizer.convert_tokens_to_ids(tgt_lang),
        max_new_tokens=256
    )

    return tokenizer.batch_decode(
        translated_tokens,
        skip_special_tokens=True
    )[0]
```

```python
print(
    translate(
        "Hello, how are you?",
        "eng_Latn",
        "hin_Deva"
    )
)
```

```python
print(
    translate(
        "Welcome to our chatbot",
        "eng_Latn",
        "tel_Telu"
    )
)
```

```python
%%writefile app.py

import streamlit as st
from langdetect import detect

# -----------------------------
# Language Dictionaries
# -----------------------------

language_names = {
    "en": "English",
    "hi": "Hindi",
    "te": "Telugu",
    "es": "Spanish"
}

greetings = {
    "en": "Hello! How can I help you today?",
    "hi": "नमस्ते! मैं आपकी कैसे सहायता कर सकता हूँ?",
    "te": "నమస్కారం! నేను మీకు ఎలా సహాయం చేయగలను?",
    "es": "¡Hola! ¿Cómo puedo ayudarte hoy?"
}

# -----------------------------
# Detect Language
# -----------------------------

def detect_language(text):
    return detect(text)

# -----------------------------
# Greeting Function
# -----------------------------

def chatbot_response(message):

    lang = detect_language(message)

    response = greetings.get(lang, greetings["en"])

    return lang, response

# -----------------------------
# Streamlit UI
# -----------------------------

st.set_page_config(page_title="Multilingual Chatbot")

st.title("🌍 Multilingual Chatbot")

message = st.text_area("Enter your message")

if st.button("Send"):

    lang, response = chatbot_response(message)

    st.subheader("Detected Language")

    st.success(language_names.get(lang, "Unknown"))

    st.subheader("Chatbot Response")

    st.info(response)
```
